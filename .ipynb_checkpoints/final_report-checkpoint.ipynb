<h1>Final Report

After cleaning up and exploring the sample dataset on insurance charges for our clients including information on their BMI, age, sex, number of children, region where care was received, and smoking status, a few noticable trends stood out.

These trends can not only allow us to make some predictions about future charges, but can also be used to guide future programs for our clients.

Below is a histogram that summarizes the overall charges for this subset of our client data. There are a significant amount of smaller charges, but we want to determine what factors, if any, are resulting in the higher charges we see:

Not surprisingly, our clients will generally experience higher charges as they age. Unfortunately, there is not much we can do to mitigate this increase except encourage them to stay as healthy and as safe as possible and continue to fund the search for the fountation of youth. The graph below shows how cost is correlated with increasing age:

Additionally, we should definitely incentivize our clients to lose weight and exercise to improve their BMI. As demonstrated below, the charges are positively correlated with BMI:

The most important finding of this analysis, however, was that whether or not our client smokes bears the most weight in determining charges.
<br><br>This will be the single most significant factor in the cost of care that they receive, as demonstrated in the graphs below which show Charges vs. Age and Charges vs. BMI with patients who smoke in orange:

As you can see, a client who is a smoker does not pay less than $10,000 for medical charges regardless of how young they might be. This presents a considerable problem and a huge cost saving opportunity. We should strongly incentivize that clients quit smoking as soon as possible.


Using our client's age, BMI, and smoker status, we built a model using linear regression that seemed to perform pretty well at predicting charges. The model was able to demonstrate that the features above likely accounted for 75% of the variation in cost of charges.
<br><br>The p-value for all features except sex were extremely small and therefore significant in determining charges.
